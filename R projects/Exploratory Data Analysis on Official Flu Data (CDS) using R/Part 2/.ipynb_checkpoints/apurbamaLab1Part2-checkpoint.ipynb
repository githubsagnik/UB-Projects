### 1) Influenza national summary (green and yellow chart) 

```r
flu_summary <- read.csv("Influenza_national_summary.csv")

library(ggplot2)
flu_summary
```

```r
# Create the data for the chart
A <- flu_summary$Total.A
B <- flu_summary$Total.B
M <- flu_summary$Week

data=data.frame(B,A)
colors <- c("green","yellow")

# Plot the bar chart 
#barplot(H,names.arg=M,xlab="Month",ylab="Revenue",col="yellow",
#main="Influenza national summary ",ylim=range(0,10000),space=0)


barplot(as.matrix(as.data.frame(t(data))),beside=F,xlab="Week",names.arg=M, col=colors,axis.lty="solid",
       ylim=c(0,10000),ylab="Number of positive specimens")
#legend("topright",rownames(t(data)),fill=colors,title="Influenza national summary")
par(new=T)
#plot(flu_summary$Percent.Positive.A,type = 'o')

plot(flu_summary$X..Positive, col="black", type="l", ylim=c(0,25), ylab="" ,xaxt="n",yaxt="n",
    xlab="")

# second y-axis
axis(4)
mtext("Percent Positive", side = 4, line = 1.35)

par(new=T)
lines(flu_summary$Percent.Positive.A, col="#FF9933", type="l", lty=2,lwd=2, ylim=c(0,25), xlab="", ylab="" ,xaxt="n",yaxt="n")
par(new=T)
lines(flu_summary$Percent.Positive.B, col="green", type="l", lty=2,lwd=2, ylim=c(0,25), xlab="", ylab="" ,xaxt="n",yaxt="n")

legend("topright", bty="n", c("A","B"), fill=c("yellow","chartreuse3"), inset=c(.1,.001))
legend("topright", bty="n", c("Percent Positive","% Positive Flu A", "% Positive Flu A"), lty=c(1,2,2),lwd=c(1,2,2), col=c("black","darkgoldenrod2","chartreuse4"), inset=c(.1,.1))
```

```r
flu_summary_52 <- read.csv("FluViewPhase2Data/WHO_NREVSS_Clinical_Labs.csv")

flu_summary_52$YEAR_WEEK <- paste(flu_summary_52$YEAR,flu_summary_52$WEEK)
flu_summary_52$YEAR_WEEK <- gsub('\\s+', '', flu_summary_52$YEAR_WEEK)
flu_summary_52
```

```r
# Create the data for the chart
A <- flu_summary_52$TOTAL.A
B <- flu_summary_52$TOTAL.B
M <- flu_summary_52$YEAR_WEEK

data=data.frame(B,A)
colors <- c("green","yellow")

# Plot the bar chart 
#barplot(H,names.arg=M,xlab="Month",ylab="Revenue",col="yellow",
#main="Influenza national summary ",ylim=range(0,10000),space=0)


barplot(as.matrix(as.data.frame(t(data))),beside=F,xlab="Week",names.arg=M, col=colors,axis.lty="solid",
       ylim=c(0,25000),ylab="Number of positive specimens")
#legend("topright",rownames(t(data)),fill=colors,title="Influenza national summary")
par(new=T)
#plot(flu_summary$Percent.Positive.A,type = 'o')

plot(flu_summary_52$PERCENT.POSITIVE, col="black", type="l", ylim=c(0,30), ylab="" ,xaxt="n",yaxt="n",
    xlab="")

# second y-axis
axis(4)
mtext("Percent Positive", side = 4, line = 1.35)

par(new=T)
lines(flu_summary_52$PERCENT.A, col="#FF9933", type="l", lty=2,lwd=2, ylim=c(0,30), xlab="", ylab="" ,xaxt="n",yaxt="n")
par(new=T)
lines(flu_summary_52$PERCENT.B, col="green", type="l", lty=2,lwd=2, ylim=c(0,30), xlab="", ylab="" ,xaxt="n",yaxt="n")

legend("topright", bty="n", c("A","B"), fill=c("yellow","chartreuse3"), inset=c(.1,.001))
legend("topright", bty="n", c("Percent Positive","% Positive Flu A", "% Positive Flu A"), lty=c(1,2,2),lwd=c(1,2,2), col=c("black","darkgoldenrod2","chartreuse4"), inset=c(.1,.1))
```
