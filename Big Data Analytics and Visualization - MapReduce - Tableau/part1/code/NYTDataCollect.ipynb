```python
import json, requests
import time
from nytimesarticle import articleAPI
from bs4 import BeautifulSoup
import requests
```

```python
query = ["gun control","gunman","nra","gun sense","firearms","second amendment"]
api = articleAPI("lpyqcLzuyzuc4yitsqpJ6Gy2Q5WBF0YA")

links=[]

try:
    for l in range(0,len(query)):
        n=0
            
        for k in range(0,10):
            
            articles = api.search(q=str(query[l]), begin_date=20190417,page=k,glocations="United States")
            for i in range(0,len(articles['response']['docs'])):
                url = articles['response']['docs'][i]['web_url']
                data = requests.get(url)
                soup = BeautifulSoup(data.content, 'html.parser')
                soup.prettify()
                filename='data/nyt/guns1daynyt0417.txt'
                f=open(filename,'a')
                for j in range((len(soup.find_all('p')))-3):
                    f.write(soup.find_all('p')[j].get_text())
                f.close()
                print(url)
                links.append(url)
                n+=1
            time.sleep(10)
        print("Number of articles for : ",query[l],len(links),"  for 2019 period 1 day")
except:
    print("Number of articles: ",len(links),"  for 2019 period 1 day")
```
